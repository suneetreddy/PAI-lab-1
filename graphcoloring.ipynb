```python
def is_safe(graph, color, v, c):
    for i in range(len(graph)):
        if graph[v][i] == 1 and color[i] == c:
            return False
    return True
solution_count = 0
def graph_coloring(graph, m, color, v):
    global solution_count
    if v == len(graph):
        solution_count += 1
        print("\nSolution", solution_count)
        for i in range(len(graph)):
            print("Vertex", i, "-> Color", color[i])
        return
    for c in range(1, m + 1):
        if is_safe(graph, color, v, c):
            color[v] = c
            graph_coloring(graph, m, color, v + 1)
            color[v] = 0
n = int(input("no of vertices: "))
m = int(input("enter colors: "))
graph = []
for i in range(n):
    row = list(map(int, input().split()))
    graph.append(row)
color = [0] * n
graph_coloring(graph, m, color, 0)
if solution_count==0:
    print("no solution")
```
