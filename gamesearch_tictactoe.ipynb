```python
import math
board = [' ']*9
def print_board():
    for i in range(0,9,3):
        print(board[i], "|", board[i+1], "|", board[i+2])
        if i < 6:
            print("--+---+--")
def winner(p):
    wins = [(0,1,2),(3,4,5),(6,7,8),
            (0,3,6),(1,4,7),(2,5,8),
            (0,4,8),(2,4,6)]
    for a,b,c in wins:
        if board[a]==board[b]==board[c]==p:
            return True
    return False
def full():
    return ' ' not in board
def alphabeta(isMax, alpha, beta):
    if winner('X'):
        return 1
    if winner('O'):
        return -1
    if full():
        return 0
    if isMax:
        best = -math.inf
        for i in range(9):
            if board[i]==' ':
                board[i]='X'
                score = alphabeta(False, alpha, beta)
                board[i]=' '
                best = max(best, score)
                alpha = max(alpha, best)
                if beta <= alpha:
                    break
        return best
    else:
        best = math.inf
        for i in range(9):
            if board[i]==' ':
                board[i]='O'
                score = alphabeta(True, alpha, beta)
                board[i]=' '
                best = min(best, score)
                beta = min(beta, best)
                if beta <= alpha:
                    break
        return best
def best_move():
    best = -math.inf
    move = -1
    for i in range(9):
        if board[i]==' ':
            board[i]='X'
            score = alphabeta(False,-math.inf,math.inf)
            board[i]=' '
            if score > best:
                best = score
                move = i
    board[move] = 'X'
while True:
    print_board()
    pos = int(input("Enter position (0-8): "))
    if board[pos]==' ':
        board[pos]='O'
    if winner('O'):
        print_board()
        print("You win")
        break
    if full():
        print_board()
        print("Draw")
        break
    best_move()
    if winner('X'):
        print_board()
        print("AI wins")
        break
    if full():
        print_board()
        print("Draw")
        break
```
