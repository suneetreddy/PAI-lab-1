```python
states = ['Interested', 'Neutral', 'Disengaged']

observations = ['low', 'medium', 'high']

start_prob = {
'Interested': 0.5,
'Neutral': 0.3,
'Disengaged': 0.2
}

transition_prob = {
'Interested': {'Interested': 0.6, 'Neutral': 0.3, 'Disengaged': 0.1},
'Neutral': {'Interested': 0.2, 'Neutral': 0.5, 'Disengaged': 0.3},
'Disengaged': {'Interested': 0.1, 'Neutral': 0.3, 'Disengaged': 0.6}
}

emission_prob = {
'Interested': {'low': 0.1, 'medium': 0.3, 'high': 0.6},
'Neutral': {'low': 0.3, 'medium': 0.5, 'high': 0.2},
'Disengaged': {'low': 0.6, 'medium': 0.3, 'high': 0.1}
}
obs_sequence = ['high', 'medium', 'low']

def forward_algorithm(states, obs_seq, start_prob, transition_prob, emission_prob):
    forward = []
    f0 = {}
    for state in states:
        f0[state] = start_prob[state] * emission_prob[state][obs_seq[0]]
    forward.append(f0)
    for t in range(1, len(obs_seq)):
        ft = {}
        for curr_state in states:
            prob = 0
            for prev_state in states:
                prob += forward[t-1][prev_state] * transition_prob[prev_state][curr_state]
            ft[curr_state] = prob * emission_prob[curr_state][obs_seq[t]]
        forward.append(ft)
    total_prob = sum(forward[-1][s] for s in states)
    return total_prob
result = forward_algorithm(
states, obs_sequence, start_prob, transition_prob, emission_prob
)
print("Probability of observation sequence:", result)
```
