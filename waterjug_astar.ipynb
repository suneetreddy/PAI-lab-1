```python
from heapq import heappush, heappop
def heuristic(state, target):
    return min(abs(state[0]-target), abs(state[1]-target))
def water_jug_astar(capA, capB, target):
    open_list = []
    visited = set()
    heappush(open_list, (0, 0, (0,0), []))
    while open_list:
        f, g, state, path = heappop(open_list)
        if state in visited:
            continue
        visited.add(state)
        path = path + [state]
        a, b = state
        if a == target or b == target:
            return path
        next_states = [
            (capA, b),
            (a, capB),
            (0, b),
            (a, 0),
            (max(0, a-(capB-b)), min(capB, b+a)),
            (min(capA, a+b), max(0, b-(capA-a)))
        ]
        for ns in next_states:
            if ns not in visited:
                g_new = g + 1
                h = heuristic(ns, target)
                heappush(open_list, (g_new+h, g_new, ns, path))
solution = water_jug_astar(4, 3, 2)
for step in solution:
    print(step)
```
