```python
sol_count = 0
def print_board(board):
    global sol_count
    sol_count += 1
    print(f"Solution Count : {sol_count}")
    for r in board:
        print(' '.join(r))
    print()
def is_valid(board, row, col, n):
    for i in range(row):
        if board[i][col] == 'Q':
            return False
    i, j = row-1, col-1
    while i >= 0 and j >= 0:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j -= 1
    i, j = row-1, col+1
    while i >= 0 and j < n:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j += 1
    return True
def place_queen(board, row, n):
    if row == n:
        print_board(board)
        return
    for col in range(n):
        if is_valid(board, row, col, n):
            board[row][col] = 'Q'
            place_queen(board, row+1, n)
            board[row][col] = '.'
n = int(input("Enter no. of queens : "))
board = [['.' for _ in range(n)] for _ in range(n)]
place_queen(board, 0, n)
```
